```python
# Heart Disease Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset (after downloading from Kaggle)
df = pd.read_csv("/kaggle/input/heart-disease-dataset/heart.csv")
```

```python
df.head()
```

```python
# View the first few rows
print("FIRST FIVE ROWS:")
print(df.head(), "\n")
```

```python
# Check dataset structure
print("DATASET INFO:")
print(df.info(), "\n")
```

```python
# Check for missing values
print("MISSING VALUES:")
print(df.isnull().sum(), "\n")
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}\n")
```

```python
# Remove duplicates if any
df = df.drop_duplicates()
```

```python
# Basic statistics
print("SUMMARY STATISTICS:")
print(df.describe(), "\n")
```

```python
# Check column data types
print("DATA TYPES:")
print(df.dtypes, "\n")
```

```python
# Unique values in categorical columns
print("UNIQUE VALUES:")
for col in ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal', 'target']:
    print(f"{col}: {df[col].unique()}")
```

```python
# Save cleaned dataset
df.to_csv("heart_cleaned.csv", index=False)
print("\n✅ Cleaned dataset saved as heart_cleaned.csv")
```

```python
# Check correlation between features
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Count plot for target variable
sns.countplot(x="target", data=df, palette="viridis")
plt.title("Heart Disease Distribution (Target)")
plt.show()
```
